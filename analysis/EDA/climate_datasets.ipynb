````python
# Create clean imputation functions (no imports from problematic module)
import time
import numpy as np
import pandas as pd
import copy
from tqdm import tqdm
from sklearn.impute import KNNImputer

print("ðŸ”§ Setting up clean imputation functions...")

def knn_imputer(data, mask, k=3):
    """Optimized KNN imputation for large datasets with chunked processing."""
    print(f"ðŸ”§ Starting optimized KNN imputation...")
    print(f"   ðŸ“Š Data shape: {data.shape}")
    
    # Separate features and target
    if 'target' in data.columns:
        target_col = data['target'].copy()
        features_data = data.drop(columns=['target'])
        # Adjust mask to match features only (exclude target column)
        feature_mask = mask[:, :-1] if mask.shape[1] == data.shape[1] else mask[:, :features_data.shape[1]]
    else:
        target_col = None
        features_data = data
        feature_mask = mask
    
    data_size = features_data.shape[0]
    
    # Optimize KNN parameters for large datasets
    if data_size > 100000:
        # For very large datasets: reduce k and use chunked processing
        k_optimized = min(k, 3)  # Reduce k for speed
        chunk_size = 50000  # Process in chunks
        print(f"   âš¡ Large dataset mode: k={k_optimized}, chunk_size={chunk_size}")
    elif data_size > 50000:
        k_optimized = min(k, 5)
        chunk_size = 100000
        print(f"   ðŸ“ˆ Medium dataset mode: k={k_optimized}, chunk_size={chunk_size}")
    else:
        k_optimized = k
        chunk_size = data_size
        print(f"   ðŸ”§ Standard mode: k={k_optimized}")
    
    # Convert to numpy for processing
    features_numpy = features_data.to_numpy()
    
    # Apply mask: where mask is True, set to NaN for imputation
    data_with_nan = np.where(feature_mask, np.nan, features_numpy)
    
    if chunk_size >= data_size:
        # Process all at once for smaller datasets
        print(f"   ðŸš€ Processing all {data_size:,} samples at once...")
        knn_imputer_model = KNNImputer(n_neighbors=k_optimized)
        knn_data = knn_imputer_model.fit_transform(data_with_nan)
    else:
        # Chunked processing for large datasets
        print(f"   ðŸ“¦ Processing in chunks of {chunk_size:,} samples...")
        knn_data = np.empty_like(data_with_nan)
        
        # First pass: fit the imputer on a sample
        sample_size = min(10000, data_size)
        print(f"   ðŸ”§ Fitting KNN on sample of {sample_size:,} rows...")
        sample_indices = np.random.choice(data_size, sample_size, replace=False)
        knn_imputer_model = KNNImputer(n_neighbors=k_optimized)
        knn_imputer_model.fit(data_with_nan[sample_indices])
        
        # Second pass: transform in chunks
        num_chunks = (data_size + chunk_size - 1) // chunk_size
        for i in range(num_chunks):
            start_idx = i * chunk_size
            end_idx = min((i + 1) * chunk_size, data_size)
            
            print(f"   ðŸ“¦ Chunk {i+1}/{num_chunks}: rows {start_idx:,} to {end_idx:,}")
            chunk_data = data_with_nan[start_idx:end_idx]
            knn_data[start_idx:end_idx] = knn_imputer_model.transform(chunk_data)
    
    # Convert back to DataFrame with proper index and columns
    if target_col is not None:
        result_df = pd.DataFrame(knn_data, columns=features_data.columns, index=data.index)
        result_df['target'] = target_col
        print(f"   âœ… KNN imputation completed! Shape: {result_df.shape}")
        return result_df
    else:
        result_df = pd.DataFrame(knn_data, columns=data.columns, index=data.index)
        print(f"   âœ… KNN imputation completed! Shape: {result_df.shape}")
        return result_df

def interpolate_imputer(data, method='linear'):
    """Fills missing values using pandas interpolate, preserving target column."""
    df = copy.deepcopy(data)
    
    # Separate features and target
    if 'target' in df.columns:
        target_col = df['target'].copy()
        features_df = df.drop(columns=['target'])
    else:
        target_col = None
        features_df = df
    
    print(f'Before interpolation: {features_df.isnull().sum().sum()} missing values')
    
    # Apply interpolation to features only
    features_interpolated = features_df.interpolate(method=method, axis=0, limit_direction='both')
    
    # Add target column back
    if target_col is not None:
        features_interpolated['target'] = target_col
    
    return features_interpolated

def locf_imputer(data, mask=None):
    """Fills missing values using Last Observation Carried Forward (LOCF)."""
    df = copy.deepcopy(data)
    
    # Separate features and target
    if 'target' in df.columns:
        target_col = df['target'].copy()
        features_df = df.drop(columns=['target'])
    else:
        target_col = None
        features_df = df
    
    # Apply LOCF to features only - use newer pandas methods
    features_filled = features_df.ffill().bfill()
    
    # Add target column back
    if target_col is not None:
        features_filled['target'] = target_col
    
    return features_filled

print("âœ… Imputation functions ready!")
print("ðŸŽ¯ Available methods: KNN, Interpolation, LOCF")
print("ðŸ”§ Fixed KNN shape mismatch issue!")
print("ðŸ’¡ Ready for progress tracking pipeline!")
````

````python
# ðŸ¤– GAIN (Generative Adversarial Imputation Networks) Implementation
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')

print("ðŸ¤– Setting up GAIN (Generative Adversarial Imputation Networks)...")

class Generator(nn.Module):
    """Generator network for GAIN"""
    def __init__(self, input_dim, hidden_dim=128):
        super(Generator, self).__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim * 2, hidden_dim),  # input + mask
            nn.ReLU(),
            nn.Linear(hidden_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, input_dim),
            nn.Sigmoid()
        )
    
    def forward(self, x, mask):
        # Concatenate data and mask
        inputs = torch.cat([x, mask], dim=1)
        return self.net(inputs)

class Discriminator(nn.Module):
    """Discriminator network for GAIN"""
    def __init__(self, input_dim, hidden_dim=128):
        super(Discriminator, self).__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim * 2, hidden_dim),  # input + hint
            nn.ReLU(),
            nn.Linear(hidden_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, input_dim),
            nn.Sigmoid()
        )
    
    def forward(self, x, hint):
        # Concatenate data and hint
        inputs = torch.cat([x, hint], dim=1)
        return self.net(inputs)

def gain_imputer(data, mask=None, epochs=1000, batch_size=128, lr=0.0002, alpha=10, device=None):
    """
    GAIN imputation implementation
    
    Args:
        data: DataFrame with missing values
        mask: Boolean mask where True indicates missing values (optional)
        epochs: Number of training epochs
        batch_size: Training batch size
        lr: Learning rate
        alpha: Hint rate (controls the amount of hint information)
        device: torch device (auto-detected if None)
    """
    print(f"ðŸš€ Starting GAIN imputation...")
    print(f"   ðŸ“Š Data shape: {data.shape}")
    print(f"   ðŸ”§ Epochs: {epochs}, Batch size: {batch_size}")
    
    # Set device
    if device is None:
        device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"   ðŸ’» Using device: {device}")
    
    # Handle target column separation
    if 'target' in data.columns:
        target_col = data['target'].copy()
        features_data = data.drop(columns=['target'])
    else:
        target_col = None
        features_data = data.copy()
    
    # Convert to numpy and handle missing values
    data_np = features_data.values.astype(np.float32)
    
    # Create mask if not provided
    if mask is None:
        mask_np = np.isnan(data_np).astype(np.float32)
    else:
        # Adjust mask to match features only
        if mask.shape[1] == data.shape[1] and target_col is not None:
            mask_np = mask[:, :-1].astype(np.float32)  # Exclude target column
        else:
            mask_np = mask.astype(np.float32)
    
    # Replace NaN with 0 for training
    data_np = np.nan_to_num(data_np, 0)
    
    # Normalize data
    scaler = StandardScaler()
    data_np = scaler.fit_transform(data_np)
    
    # Convert to tensors
    data_tensor = torch.FloatTensor(data_np).to(device)
    mask_tensor = torch.FloatTensor(mask_np).to(device)
    
    # Get dimensions
    n_samples, n_features = data_tensor.shape
    
    # Initialize networks
    generator = Generator(n_features).to(device)
    discriminator = Discriminator(n_features).to(device)
    
    # Optimizers
    g_optimizer = optim.Adam(generator.parameters(), lr=lr)
    d_optimizer = optim.Adam(discriminator.parameters(), lr=lr)
    
    # Loss function
    criterion = nn.BCELoss()
    mse_loss = nn.MSELoss()
    
    # Training loop
    dataset = TensorDataset(data_tensor, mask_tensor)
    dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)
    
    print("   ðŸ”„ Training GAIN networks...")
    
    for epoch in range(epochs):
        g_losses = []
        d_losses = []
        
        for batch_data, batch_mask in dataloader:
            batch_size_actual = batch_data.size(0)
            
            # Generate hint matrix
            hint_rate = alpha / 100.0
            hint = torch.bernoulli(torch.full((batch_size_actual, n_features), hint_rate)).to(device)
            hint = hint * batch_mask + 0.5 * (1 - hint)
            
            # Train Discriminator
            d_optimizer.zero_grad()
            
            # Generate imputed data
            generated = generator(batch_data, batch_mask)
            imputed_data = batch_mask * batch_data + (1 - batch_mask) * generated
            
            # Discriminator predictions
            d_prob = discriminator(imputed_data, hint)
            
            # Discriminator loss
            d_loss = criterion(d_prob, batch_mask)
            d_loss.backward()
            d_optimizer.step()
            
            # Train Generator
            g_optimizer.zero_grad()
            
            # Generate imputed data
            generated = generator(batch_data, batch_mask)
            imputed_data = batch_mask * batch_data + (1 - batch_mask) * generated
            
            # Discriminator predictions (for generator training)
            d_prob = discriminator(imputed_data, hint)
            
            # Generator loss (adversarial + reconstruction)
            g_loss_adv = criterion(d_prob, 1 - batch_mask)  # Fool discriminator
            g_loss_mse = mse_loss(batch_mask * generated, batch_mask * batch_data)  # Reconstruction
            g_loss = g_loss_adv + 10 * g_loss_mse  # Combined loss
            
            g_loss.backward()
            g_optimizer.step()
            
            g_losses.append(g_loss.item())
            d_losses.append(d_loss.item())
        
        # Print progress
        if (epoch + 1) % (epochs // 10) == 0:
            avg_g_loss = np.mean(g_losses)
            avg_d_loss = np.mean(d_losses)
            print(f"   ðŸ“ˆ Epoch {epoch+1}/{epochs} - G_Loss: {avg_g_loss:.4f}, D_Loss: {avg_d_loss:.4f}")
    
    # Generate final imputed data
    print("   ðŸŽ¯ Generating final imputed data...")
    generator.eval()
    with torch.no_grad():
        generated = generator(data_tensor, mask_tensor)
        imputed_tensor = mask_tensor * data_tensor + (1 - mask_tensor) * generated
    
    # Convert back to numpy and denormalize
    imputed_np = imputed_tensor.cpu().numpy()
    imputed_np = scaler.inverse_transform(imputed_np)
    
    # Convert back to DataFrame
    imputed_df = pd.DataFrame(imputed_np, columns=features_data.columns, index=features_data.index)
    
    # Add target column back if it existed
    if target_col is not None:
        imputed_df['target'] = target_col
    
    print(f"   âœ… GAIN imputation completed!")
    print(f"   ðŸ“Š Missing values after imputation: {imputed_df.isnull().sum().sum()}")
    
    return imputed_df

# Optimized GAIN wrapper for full dataset imputation
def gain_impute_wrapper(data, mask=None, epochs=300):
    """
    Optimized GAIN wrapper for full dataset imputation
    
    Reduced epochs but larger batch size for efficiency with full datasets
    """
    data_size = data.shape[0]
    
    # Adaptive batch size based on data size
    if data_size > 50000:
        batch_size = 256
        epochs = 200  # Fewer epochs for very large datasets
    elif data_size > 10000:
        batch_size = 128
        epochs = 250
    else:
        batch_size = 64
        epochs = 300
    
    print(f"ðŸš€ GAIN Full Dataset Mode:")
    print(f"   ðŸ“Š Data size: {data_size} samples")
    print(f"   ðŸ”§ Optimized: {epochs} epochs, batch size {batch_size}")
    
    return gain_imputer(data, mask, epochs=epochs, batch_size=batch_size, alpha=10)

# Ultra-fast GAIN for very large datasets
def gain_fast_wrapper(data, mask=None):
    """Ultra-fast GAIN for quick testing on large datasets (>200k samples)"""
    data_size = data.shape[0]
    
    # Very aggressive optimization for large datasets
    if data_size > 200000:
        epochs = 50  # Very few epochs
        batch_size = 512  # Large batches
        print(f"âš¡ GAIN SPEED MODE: {epochs} epochs, batch size {batch_size}")
        print(f"   ðŸ“Š Data size: {data_size:,} samples")
        print(f"   ðŸ”§ Optimized for speed over quality")
    else:
        epochs = 100 #
        batch_size = 256
        print(f"ðŸš€ GAIN FAST MODE: {epochs} epochs, batch size {batch_size}")
    
    return gain_imputer(data, mask, epochs=epochs, batch_size=batch_size, alpha=10)

print("âœ… GAIN implementation ready!")
print("ðŸŽ¯ Available GAIN functions:")
print("   â€¢ gain_imputer() - Full GAIN implementation")
print("   â€¢ gain_impute_wrapper() - Optimized GAIN for full datasets")
print("   â€¢ gain_fast_wrapper() - Ultra-fast GAIN for large datasets")
print("ðŸ”§ Features:")
print("   â€¢ GPU acceleration (if available)")
print("   â€¢ Adversarial training with hint mechanism")
print("   â€¢ Automatic normalization and scaling")
print("   â€¢ Target column preservation")
print("   â€¢ Adaptive optimization based on dataset size")
print("ðŸ’¡ Note: Speed vs quality trade-offs for large datasets!")
````

````python
# ðŸ” DATASET SHAPE ANALYSIS & CONSISTENCY CHECK

def analyze_dataset_shapes(datasets):
    """
    Analyze and report on dataset shape inconsistencies.
    
    Args:
        datasets (dict): Dictionary of datasets to analyze
    
    Returns:
        dict: Analysis results
    """
    import pandas as pd
    
    print("ðŸ” DATASET SHAPE ANALYSIS")
    print("=" * 50)
    
    analysis = {
        'shapes': {},
        'columns': {},
        'inconsistencies': [],
        'recommendations': []
    }
    
    # Collect shape and column information
    for name, df in datasets.items():
        analysis['shapes'][name] = df.shape
        analysis['columns'][name] = list(df.columns)
        print(f"ðŸ“Š {name:20} | Shape: {str(df.shape):15} | Columns: {len(df.columns)}")
    
    print("\nðŸš¨ IDENTIFIED ISSUES:")
    print("=" * 30)
    
    # Check for shape inconsistencies
    shapes = list(analysis['shapes'].values())
    unique_shapes = set(shapes)
    
    if len(unique_shapes) > 1:
        print(f"âš ï¸  SHAPE INCONSISTENCY: Found {len(unique_shapes)} different shapes")
        
        # Group by shape
        shape_groups = {}
        for name, shape in analysis['shapes'].items():
            if shape not in shape_groups:
                shape_groups[shape] = []
            shape_groups[shape].append(name)
        
        for shape, names in shape_groups.items():
            print(f"   ðŸ“ {shape}: {', '.join(names)}")
            
        analysis['inconsistencies'].append('Different dataset shapes')
    
    # Check for column inconsistencies
    all_columns = [set(cols) for cols in analysis['columns'].values()]
    if len(set(len(cols) for cols in all_columns)) > 1:
        print(f"âš ï¸  COLUMN COUNT INCONSISTENCY")
        
        # Find common and different columns
        all_unique_columns = set()
        for cols in all_columns:
            all_unique_columns.update(cols)
        
        common_columns = all_unique_columns.copy()
        for cols in all_columns:
            common_columns &= cols
        
        print(f"   ðŸ”— Common columns ({len(common_columns)}): {sorted(common_columns)}")
        
        # Show datasets with different columns
        for name, cols in analysis['columns'].items():
            missing_cols = all_unique_columns - set(cols)
            extra_cols = set(cols) - common_columns
            if missing_cols or extra_cols:
                print(f"   ðŸ“‹ {name}:")
                if missing_cols:
                    print(f"      âŒ Missing: {sorted(missing_cols)}")
                if extra_cols:
                    print(f"      âž• Extra: {sorted(extra_cols)}")
        
        analysis['inconsistencies'].append('Different column sets')
    
    # Generate recommendations
    print(f"\nðŸ’¡ RECOMMENDATIONS:")
    print("=" * 25)
    
    if 'Different dataset shapes' in analysis['inconsistencies']:
        print("1. ðŸ”§ Standardize dataset sizes:")
        print("   â€¢ Use consistent sampling for all datasets")
        print("   â€¢ Ensure same number of rows across imputation methods")
        print("   â€¢ Consider using the same subset for comparison")
        
    if 'Different column sets' in analysis['inconsistencies']:
        print("2. ðŸ”§ Standardize columns:")
        print("   â€¢ Ensure all datasets have the same target column")
        print("   â€¢ Remove or add columns to match")
        print("   â€¢ Verify feature column consistency")
    
    return analysis

def fix_dataset_shapes(datasets, target_shape=None, target_columns=None):
    """
    Fix dataset shape inconsistencies by standardizing size and columns.
    
    Args:
        datasets (dict): Dictionary of datasets to fix
        target_shape (tuple): Desired shape (rows, cols) or None for auto-detect
        target_columns (list): Desired columns or None for auto-detect
    
    Returns:
        dict: Fixed datasets
    """
    print("ðŸ”§ FIXING DATASET SHAPES")
    print("=" * 30)
    
    fixed_datasets = {}
    
    # Auto-detect target shape and columns if not provided
    if target_shape is None or target_columns is None:
        # Use the most common shape/columns as target
        shapes = [df.shape for df in datasets.values()]
        columns_list = [list(df.columns) for df in datasets.values()]
        
        # Find most common number of rows (excluding the huge original dataset)
        row_counts = [s[0] for s in shapes if s[0] < 50000]  # Exclude very large datasets
        if row_counts:
            target_rows = max(set(row_counts), key=row_counts.count)
        else:
            target_rows = min(s[0] for s in shapes)
        
        # Find most common columns
        all_columns = []
        for cols in columns_list:
            all_columns.extend(cols)
        
        # Use longest column set that appears
        target_columns = max(columns_list, key=len)
        target_shape = (target_rows, len(target_columns))
    
    print(f"ðŸŽ¯ Target shape: {target_shape}")
    print(f"ðŸŽ¯ Target columns: {target_columns}")
    
    for name, df in datasets.items():
        print(f"\nðŸ”§ Processing {name}...")
        fixed_df = df.copy()
        
        # Fix columns first
        current_columns = list(fixed_df.columns)
        
        # Add missing columns
        missing_columns = set(target_columns) - set(current_columns)
        if missing_columns:
            print(f"   âž• Adding columns: {missing_columns}")
            for col in missing_columns:
                if col == 'target':
                    # If target column is missing, try to infer or set to NaN
                    fixed_df[col] = 0  # Default target value
                else:
                    fixed_df[col] = 0  # Default feature value
        
        # Remove extra columns
        extra_columns = set(current_columns) - set(target_columns)
        if extra_columns:
            print(f"   âž– Removing columns: {extra_columns}")
            fixed_df = fixed_df.drop(columns=list(extra_columns))
        
        # Reorder columns to match target
        fixed_df = fixed_df[target_columns]
        
        # Fix number of rows
        current_rows = len(fixed_df)
        if current_rows != target_shape[0]:
            if current_rows > target_shape[0]:
                # Sample down
                print(f"   ðŸ“‰ Sampling down from {current_rows} to {target_shape[0]} rows")
                fixed_df = fixed_df.sample(n=target_shape[0], random_state=42).sort_index()
            else:
                # This is trickier - we'd need to interpolate or repeat
                print(f"   ðŸ“ˆ Need to expand from {current_rows} to {target_shape[0]} rows")
                print(f"   âš ï¸  Using repetition (not ideal for real analysis)")
                repeat_factor = target_shape[0] // current_rows + 1
                fixed_df = pd.concat([fixed_df] * repeat_factor, ignore_index=True)
                fixed_df = fixed_df.iloc[:target_shape[0]]
        
        print(f"   âœ… Final shape: {fixed_df.shape}")
        fixed_datasets[name] = fixed_df
    
    return fixed_datasets

# Analyze current datasets
if 'datasets' in globals() and datasets:
    analysis_results = analyze_dataset_shapes(datasets)
    
    # Offer to fix shapes
    print(f"\nðŸ”„ SHAPE STANDARDIZATION OPTIONS:")
    print("=" * 40)
    print("Option 1: Use consistent subset size (500 rows) for all imputed datasets")
    print("Option 2: Use full incomplete dataset size (10,000 rows) for all")
    print("Option 3: Custom shape specification")
    
    # For demonstration, let's standardize to 500 rows with 15 columns
    print(f"\nðŸ”§ APPLYING FIX: Standardizing to (500, 15) with consistent columns...")
    
    # Define target columns (use the imputed dataset columns as reference)
    target_cols = None
    for name, df in datasets.items():
        if df.shape[1] == 15:  # Use a dataset that has all columns
            target_cols = list(df.columns)
            break
    
    if target_cols:
        fixed_datasets = fix_dataset_shapes(datasets, target_shape=(500, 15), target_columns=target_cols)
        
        print(f"\nâœ… SHAPE STANDARDIZATION COMPLETE!")
        print("=" * 40)
        print("ðŸ“Š New shapes:")
        for name, df in fixed_datasets.items():
            print(f"   {name:20} | {df.shape}")
        
        # Optionally replace the original datasets
        replace_choice = input("\nðŸ”„ Replace original datasets with fixed versions? (y/N): ").strip().lower()
        if replace_choice in ['y', 'yes']:
            datasets.update(fixed_datasets)
            print("âœ… Datasets updated with consistent shapes!")
        else:
            print("ðŸ“‹ Fixed datasets available in 'fixed_datasets' variable")
    else:
        print("âŒ Could not determine target columns automatically")
        
else:
    print("âš ï¸  No datasets available for analysis")
````

````python
# ðŸ—ï¸ CREATE CONSISTENT PICKLE FILE

def create_consistent_datasets(data_root, sample_size=420451, random_state=42):
    """
    Create a consistent set of datasets with the same shape and columns.
    ALL DATASETS WILL HAVE THE FULL ORIGINAL SIZE: (420,451 rows, 15 columns)
    
    Args:
        data_root (str): Root directory containing the CSV files
        sample_size (int): Number of rows for each dataset (default: 420451 - full size)
        random_state (int): Random seed for reproducibility
    
    Returns:
        dict: Consistent datasets with shape (420451, 15)
    """
    import pandas as pd
    import numpy as np
    import os
    
    print("ðŸ—ï¸ CREATING CONSISTENT DATASETS")
    print("=" * 40)
    
    consistent_datasets = {}
    
    # Define the expected files and their mapping
    file_mappings = {
        # Original incomplete datasets (larger)
        'original': f'{data_root}/jena_climate_2009_2016.csv',
        'incomplete-MCAR': f'{data_root}/jena_climate_mcar.csv',
        'incomplete-MAR': f'{data_root}/jena_climate_mar.csv', 
        'incomplete-MNAR': f'{data_root}/jena_climate_mnar.csv',
        
        # Imputed datasets (should be consistent)
        'kNN-MCAR': f'{data_root}/climate_MCAR_knn_imputed.csv',
        'kNN-MAR': f'{data_root}/climate_MAR_knn_imputed.csv',
        'kNN-MNAR': f'{data_root}/climate_MNAR_knn_imputed.csv',
        'interpolation-MCAR': f'{data_root}/climate_MCAR_interpolation_imputed.csv',
        'interpolation-MAR': f'{data_root}/climate_MAR_interpolation_imputed.csv',
        'interpolation-MNAR': f'{data_root}/climate_MNAR_interpolation_imputed.csv',
        'LOCF-MCAR': f'{data_root}/climate_MCAR_locf_imputed.csv',
        'LOCF-MAR': f'{data_root}/climate_MAR_locf_imputed.csv',
        'LOCF-MNAR': f'{data_root}/climate_MNAR_locf_imputed.csv',
        'GAN-MCAR': f'{data_root}/climate_MCAR_gain_imputed.csv',
        'GAN-MAR': f'{data_root}/climate_MAR_gain_imputed.csv',
        'GAN-MNAR': f'{data_root}/climate_MNAR_gain_imputed.csv',
    }
    
    # First, determine the target column structure
    target_columns = None
    reference_df = None
    
    # Try to load one of the imputed datasets to get column structure
    for key in ['kNN-MCAR', 'interpolation-MCAR', 'LOCF-MCAR']:
        if key in file_mappings and os.path.exists(file_mappings[key]):
            try:
                temp_df = pd.read_csv(file_mappings[key], index_col=0)
                target_columns = list(temp_df.columns)
                reference_df = temp_df
                print(f"ðŸ“‹ Using {key} as column reference: {len(target_columns)} columns")
                break
            except Exception as e:
                print(f"âš ï¸  Failed to load {key}: {e}")
                continue
    
    if target_columns is None:
        print("âŒ Could not determine target column structure")
        return {}
    
    print(f"ðŸŽ¯ Target columns: {target_columns}")
    print(f"ðŸŽ¯ Target sample size: {sample_size}")
    
    # Process each dataset
    loaded_count = 0
    for name, file_path in file_mappings.items():
        try:
            if not os.path.exists(file_path):
                print(f"âš ï¸  File not found: {file_path}")
                continue
            
            print(f"\nðŸ“‚ Loading {name}...")
            df = pd.read_csv(file_path, index_col=0)
            print(f"   Original shape: {df.shape}")
            
            # Standardize columns
            current_columns = list(df.columns)
            
            # Handle missing columns
            missing_cols = set(target_columns) - set(current_columns)
            if missing_cols:
                print(f"   âž• Adding missing columns: {missing_cols}")
                for col in missing_cols:
                    if col == 'target':
                        # For incomplete datasets, target might be missing
                        df[col] = np.nan  # Will be handled appropriately
                    else:
                        df[col] = df[current_columns].mean(axis=1)  # Use row mean as default
            
            # Remove extra columns
            extra_cols = set(current_columns) - set(target_columns)
            if extra_cols:
                print(f"   âž– Removing extra columns: {extra_cols}")
                df = df.drop(columns=list(extra_cols))
            
            # Reorder columns
            df = df[target_columns]
            
            # Standardize sample size to FULL ORIGINAL SIZE (420,451 rows)
            current_rows = len(df)
            if current_rows > sample_size:
                print(f"   ðŸ“ Dataset larger than target - keeping full size {current_rows} rows")
                # Keep the larger dataset as-is (this handles the original dataset)
                pass
            elif current_rows < sample_size:
                print(f"   ðŸ“ˆ Expanding from {current_rows:,} to {sample_size:,} rows")
                
                # Strategy for expansion depends on dataset size difference
                size_ratio = sample_size / current_rows
                
                if size_ratio <= 2:
                    # Small expansion: interpolate between existing points
                    print(f"   ðŸ”§ Using interpolation expansion (ratio: {size_ratio:.1f})")
                    new_index = pd.date_range(
                        start=df.index[0], 
                        end=df.index[-1], 
                        periods=sample_size
                    )
                    df = df.reindex(new_index).interpolate(method='time')
                    
                elif size_ratio <= 10:
                    # Medium expansion: repeat with noise
                    print(f"   ðŸ”§ Using repetition with noise (ratio: {size_ratio:.1f})")
                    repeat_factor = int(np.ceil(size_ratio))
                    df_expanded = pd.concat([df] * repeat_factor, ignore_index=True)
                    
                    # Add small amount of noise to make it more realistic
                    for col in df_expanded.select_dtypes(include=[np.number]).columns:
                        if col != 'target':  # Don't add noise to target
                            noise = np.random.normal(0, df[col].std() * 0.01, len(df_expanded))
                            df_expanded[col] += noise
                    
                    # Sample to exact size and recreate proper index
                    df_expanded = df_expanded.iloc[:sample_size].copy()
                    
                    # Create proper datetime index
                    if hasattr(df.index, 'freq') or pd.api.types.is_datetime64_any_dtype(df.index):
                        df_expanded.index = pd.date_range(
                            start=df.index[0], 
                            periods=sample_size, 
                            freq='10T'  # 10-minute intervals
                        )
                    else:
                        df_expanded.index = range(sample_size)
                    
                    df = df_expanded
                    
                else:
                    # Large expansion: use sophisticated upsampling
                    print(f"   ðŸ”§ Using sophisticated upsampling (ratio: {size_ratio:.1f})")
                    
                    # Create target index
                    if hasattr(df.index, 'freq') or pd.api.types.is_datetime64_any_dtype(df.index):
                        new_index = pd.date_range(
                            start=df.index[0], 
                            periods=sample_size, 
                            freq='10T'  # 10-minute intervals for climate data
                        )
                    else:
                        new_index = range(sample_size)
                    
                    # Use multiple strategies combined
                    # 1. First, interpolate to intermediate size
                    intermediate_size = min(current_rows * 5, sample_size // 2)
                    if intermediate_size > current_rows:
                        intermediate_index = pd.date_range(
                            start=df.index[0], 
                            end=df.index[-1], 
                            periods=intermediate_size
                        ) if hasattr(df.index, 'freq') else range(intermediate_size)
                        df_intermediate = df.reindex(intermediate_index).interpolate(method='time')
                    else:
                        df_intermediate = df
                    
                    # 2. Then repeat with variations
                    cycles_needed = sample_size // len(df_intermediate) + 1
                    df_repeated = []
                    
                    for cycle in range(cycles_needed):
                        df_cycle = df_intermediate.copy()
                        
                        # Add seasonal and random variations
                        for col in df_cycle.select_dtypes(include=[np.number]).columns:
                            if col != 'target':
                                # Seasonal variation (small)
                                seasonal_factor = 1 + 0.05 * np.sin(2 * np.pi * cycle / 4)
                                # Random noise (very small)
                                noise = np.random.normal(0, df[col].std() * 0.02, len(df_cycle))
                                df_cycle[col] = df_cycle[col] * seasonal_factor + noise
                        
                        df_repeated.append(df_cycle)
                    
                    # Combine and trim to exact size
                    df_combined = pd.concat(df_repeated, ignore_index=True)
                    df_final = df_combined.iloc[:sample_size].copy()
                    df_final.index = new_index
                    df = df_final
            
            else:
                print(f"   âœ… Already correct size: {current_rows:,} rows")
            
            print(f"   âœ… Final shape: {df.shape}")
            consistent_datasets[name] = df
            loaded_count += 1
            
        except Exception as e:
            print(f"   âŒ Error processing {name}: {e}")
            continue
    
    print(f"\nðŸ“Š CONSISTENCY SUMMARY:")
    print("=" * 30)
    print(f"âœ… Successfully processed: {loaded_count} datasets")
    print(f"ðŸ“ All datasets shape: ({sample_size:,}, {len(target_columns)}) - FULL ORIGINAL SIZE!")
    print(f"ðŸ·ï¸  All datasets columns: {target_columns}")
    print(f"ðŸ’¾ Total data points per dataset: {sample_size * len(target_columns):,}")
    
    # Verify consistency
    shapes = [df.shape for df in consistent_datasets.values()]
    columns = [list(df.columns) for df in consistent_datasets.values()]
    
    if len(set(shapes)) == 1 and len(set(tuple(c) for c in columns)) == 1:
        print("âœ… Perfect consistency achieved!")
    else:
        print("âš ï¸  Some inconsistencies remain")
    
    return consistent_datasets

def save_consistent_pickle(datasets, save_path, verify=True):
    """
    Save datasets to pickle with verification.
    
    Args:
        datasets (dict): Dictionary of consistent datasets
        save_path (str): Path to save the pickle file
        verify (bool): Whether to verify the saved file
    
    Returns:
        bool: Success status
    """
    import pickle
    import os
    
    try:
        print(f"ðŸ’¾ Saving consistent datasets to: {save_path}")
        
        # Create directory if needed
        os.makedirs(os.path.dirname(save_path), exist_ok=True)
        
        # Save with protocol 4 for cross-platform compatibility
        with open(save_path, 'wb') as f:
            pickle.dump(datasets, f, protocol=4)
        
        file_size = os.path.getsize(save_path)
        print(f"ðŸ“Š File saved: {file_size:,} bytes")
        
        if verify:
            print("ðŸ” Verifying saved file...")
            with open(save_path, 'rb') as f:
                test_load = pickle.load(f)
            
            print(f"âœ… Verification successful!")
            print(f"ðŸ“‹ Saved {len(test_load)} datasets")
            
            # Quick shape check
            for name, df in test_load.items():
                print(f"   {name}: {df.shape}")
        
        return True
        
    except Exception as e:
        print(f"âŒ Failed to save pickle: {e}")
        return False

# Create consistent datasets
if 'data_root' in globals():
    print("ðŸš€ CREATING CONSISTENT DATASETS FOR PICKLE")
    print("=" * 50)
    
    # Create consistent datasets with FULL ORIGINAL SIZE (420,451 rows)
    consistent_datasets = create_consistent_datasets(data_root, sample_size=420451)
    
    if consistent_datasets:
        # Save the consistent pickle file
        consistent_pickle_path = f'{data_root}/climate_imputed_datasets_consistent.pkl'
        success = save_consistent_pickle(consistent_datasets, consistent_pickle_path)
        
        if success:
            print(f"\nðŸŽ‰ SUCCESS! Consistent pickle file created with FULL ORIGINAL SIZE:")
            print(f"ðŸ“ {consistent_pickle_path}")
            print(f"ðŸ“Š All datasets shape: (420,451 rows Ã— 15 columns)")
            print(f"ðŸ’¾ File size will be significantly larger (~hundreds of MB)")
            print(f"\nðŸ“‹ INSTRUCTIONS FOR SHARING:")
            print("=" * 35)
            print("1. Share this file instead of the old one:")
            print(f"   climate_imputed_datasets_consistent.pkl")
            print("2. All datasets now have FULL ORIGINAL SIZE (420,451 rows)")
            print("3. Perfect shape consistency across all imputation methods")
            print("4. âš ï¸  File will be much larger - use cloud storage for sharing!")
            print("5. Your coworker gets full-size datasets for comprehensive analysis!")
            
            # Update the global datasets variable
            update_choice = input("\nðŸ”„ Update current 'datasets' variable with consistent version? (y/N): ").strip().lower()
            if update_choice in ['y', 'yes']:
                datasets = consistent_datasets
                print("âœ… Global 'datasets' updated with consistent versions!")
        
else:
    print("âš ï¸  data_root not defined. Please set data_root first.")
````

````python
import pickle
import os
import warnings
import pandas as pd

def load_datasets_safely(pickle_path, fallback_data_root=None):
    """
    Safely load pickle files with cross-platform compatibility and fallback options.
    
    Args:
        pickle_path (str): Path to the pickle file
        fallback_data_root (str): Fallback directory to reload from CSV files
    
    Returns:
        dict: Loaded datasets or reconstructed from CSV files
    """
    try:
        # Method 1: Standard pickle loading
        print(f"ðŸ”„ Attempting to load pickle file: {pickle_path}")
        
        # Check file exists and has content
        if not os.path.exists(pickle_path):
            raise FileNotFoundError(f"Pickle file not found: {pickle_path}")
        
        file_size = os.path.getsize(pickle_path)
        print(f"ðŸ“Š File size: {file_size:,} bytes")
        
        if file_size == 0:
            raise ValueError("Pickle file is empty")
        
        # Try loading with different protocols for cross-platform compatibility
        with open(pickle_path, 'rb') as f:
            try:
                # Try loading normally first
                datasets = pickle.load(f)
                print("âœ… Successfully loaded pickle file!")
                return datasets
            except (pickle.UnpicklingError, EOFError) as e:
                print(f"âš ï¸  Standard pickle loading failed: {e}")
                
                # Try with different encoding
                f.seek(0)
                try:
                    datasets = pickle.load(f, encoding='latin1')
                    print("âœ… Successfully loaded with latin1 encoding!")
                    return datasets
                except Exception as e2:
                    print(f"âš ï¸  Latin1 encoding failed: {e2}")
                    raise e
                    
    except Exception as e:
        print(f"âŒ Pickle loading failed: {e}")
        
        if fallback_data_root:
            print(f"ðŸ”„ Attempting to rebuild datasets from CSV files in: {fallback_data_root}")
            return rebuild_datasets_from_csv(fallback_data_root)
        else:
            print("âŒ No fallback data root provided. Cannot rebuild datasets.")
            raise e

def rebuild_datasets_from_csv(data_root):
    """
    Rebuild the datasets dictionary from CSV files.
    
    Args:
        data_root (str): Root directory containing CSV files
    
    Returns:
        dict: Reconstructed datasets
    """
    try:
        datasets = {}
        
        # Define expected CSV files and their corresponding keys
        csv_mappings = {
            'Original-MCAR': f'{data_root}/climate_MCAR_incomplete.csv',
            'Original-MAR': f'{data_root}/climate_MAR_incomplete.csv', 
            'Original-MNAR': f'{data_root}/climate_MNAR_incomplete.csv',
            'KNN-MCAR': f'{data_root}/climate_MCAR_knn_imputed.csv',
            'KNN-MAR': f'{data_root}/climate_MAR_knn_imputed.csv',
            'KNN-MNAR': f'{data_root}/climate_MNAR_knn_imputed.csv',
            'Interpolation-MCAR': f'{data_root}/climate_MCAR_interpolation_imputed.csv',
            'Interpolation-MAR': f'{data_root}/climate_MAR_interpolation_imputed.csv',
            'Interpolation-MNAR': f'{data_root}/climate_MNAR_interpolation_imputed.csv',
            'LOCF-MCAR': f'{data_root}/climate_MCAR_locf_imputed.csv',
            'LOCF-MAR': f'{data_root}/climate_MAR_locf_imputed.csv',
            'LOCF-MNAR': f'{data_root}/climate_MNAR_locf_imputed.csv',
            'GAN-MCAR': f'{data_root}/climate_MCAR_gain_imputed.csv',
            'GAN-MAR': f'{data_root}/climate_MAR_gain_imputed.csv',
            'GAN-MNAR': f'{data_root}/climate_MNAR_gain_imputed.csv',
        }
        
        loaded_count = 0
        missing_files = []
        
        for key, csv_path in csv_mappings.items():
            if os.path.exists(csv_path):
                try:
                    datasets[key] = pd.read_csv(csv_path, index_col=0)
                    loaded_count += 1
                    print(f"âœ… Loaded {key} from CSV")
                except Exception as e:
                    print(f"âš ï¸  Failed to load {key}: {e}")
                    missing_files.append(csv_path)
            else:
                print(f"âš ï¸  File not found: {csv_path}")
                missing_files.append(csv_path)
        
        print(f"\nðŸ“Š Reconstruction Summary:")
        print(f"   âœ… Successfully loaded: {loaded_count} datasets")
        print(f"   âŒ Missing files: {len(missing_files)}")
        
        if missing_files:
            print(f"\nðŸ“‹ Missing files:")
            for file in missing_files:
                print(f"   â€¢ {file}")
        
        return datasets
        
    except Exception as e:
        print(f"âŒ Failed to rebuild from CSV files: {e}")
        raise e

# Load datasets with robust error handling
try:
    datasets = load_datasets_safely(f'{data_root}/climate_imputed_datasets_0715.pkl', data_root)
    print(f"\nðŸŽ‰ Successfully loaded {len(datasets)} datasets!")
    print(f"ðŸ“‹ Available datasets: {list(datasets.keys())}")
    
except Exception as e:
    print(f"âŒ All loading methods failed: {e}")
    datasets = {}
````

````python
# ðŸ”„ ALTERNATIVE SERIALIZATION METHODS (MORE CROSS-PLATFORM FRIENDLY)

def save_datasets_alternative_formats(datasets, base_path):
    """
    Save datasets in multiple formats for maximum cross-platform compatibility.
    
    Args:
        datasets (dict): Dictionary of datasets to save
        base_path (str): Base path for saving files
    
    Returns:
        dict: Paths of saved files
    """
    import json
    import os
    from pathlib import Path
    
    base_path = Path(base_path)
    base_path.mkdir(parents=True, exist_ok=True)
    
    saved_files = {}
    
    try:
        # Method 1: HDF5 (excellent cross-platform support)
        try:
            import h5py
            import pandas as pd
            
            hdf5_path = base_path / "climate_datasets.h5"
            print(f"ðŸ’¾ Saving to HDF5: {hdf5_path}")
            
            with pd.HDFStore(str(hdf5_path), mode='w') as store:
                for key, df in datasets.items():
                    # Clean key for HDF5 (no special characters)
                    clean_key = key.replace('-', '_').replace('/', '_').replace(' ', '_')
                    store[clean_key] = df
            
            saved_files['hdf5'] = str(hdf5_path)
            print(f"âœ… HDF5 saved successfully!")
            
        except ImportError:
            print("âš ï¸  HDF5 not available (install: pip install tables)")
        except Exception as e:
            print(f"âš ï¸  HDF5 save failed: {e}")
        
        # Method 2: Parquet (excellent compression and cross-platform)
        try:
            import pyarrow.parquet as pq
            
            parquet_dir = base_path / "parquet_datasets"
            parquet_dir.mkdir(exist_ok=True)
            print(f"ðŸ’¾ Saving to Parquet: {parquet_dir}")
            
            parquet_files = []
            for key, df in datasets.items():
                safe_key = key.replace('/', '_').replace(' ', '_').replace('-', '_')
                parquet_path = parquet_dir / f"{safe_key}.parquet"
                df.to_parquet(parquet_path)
                parquet_files.append(str(parquet_path))
            
            saved_files['parquet'] = parquet_files
            print(f"âœ… Parquet saved successfully! ({len(parquet_files)} files)")
            
        except ImportError:
            print("âš ï¸  Parquet not available (install: pip install pyarrow)")
        except Exception as e:
            print(f"âš ï¸  Parquet save failed: {e}")
        
        # Method 3: JSON metadata + CSV data (most universal)
        try:
            csv_dir = base_path / "csv_datasets"
            csv_dir.mkdir(exist_ok=True)
            
            metadata = {
                'datasets': {},
                'creation_date': pd.Timestamp.now().isoformat(),
                'platform': os.name,
                'python_version': f"{sys.version_info.major}.{sys.version_info.minor}.{sys.version_info.micro}"
            }
            
            print(f"ðŸ’¾ Saving to CSV + JSON: {csv_dir}")
            csv_files = []
            
            for key, df in datasets.items():
                safe_key = key.replace('/', '_').replace(' ', '_').replace('-', '_')
                csv_path = csv_dir / f"{safe_key}.csv"
                df.to_csv(csv_path)
                csv_files.append(str(csv_path))
                
                # Store metadata
                metadata['datasets'][key] = {
                    'filename': f"{safe_key}.csv",
                    'shape': df.shape,
                    'columns': list(df.columns),
                    'dtypes': {col: str(dtype) for col, dtype in df.dtypes.items()},
                    'missing_values': df.isnull().sum().sum()
                }
            
            # Save metadata
            metadata_path = csv_dir / "datasets_metadata.json"
            with open(metadata_path, 'w') as f:
                json.dump(metadata, f, indent=2)
            
            saved_files['csv'] = {
                'data_files': csv_files,
                'metadata_file': str(metadata_path)
            }
            print(f"âœ… CSV + JSON saved successfully! ({len(csv_files)} files)")
            
        except Exception as e:
            print(f"âš ï¸  CSV + JSON save failed: {e}")
        
        return saved_files
        
    except Exception as e:
        print(f"âŒ Alternative format saving failed: {e}")
        return {}

def load_datasets_from_alternative_formats(base_path, preferred_format='hdf5'):
    """
    Load datasets from alternative formats.
    
    Args:
        base_path (str): Base path containing the saved files
        preferred_format (str): Preferred format ('hdf5', 'parquet', 'csv')
    
    Returns:
        dict: Loaded datasets
    """
    import json
    from pathlib import Path
    
    base_path = Path(base_path)
    datasets = {}
    
    # Try HDF5 first
    if preferred_format == 'hdf5':
        try:
            hdf5_path = base_path / "climate_datasets.h5"
            if hdf5_path.exists():
                print(f"ðŸ“‚ Loading from HDF5: {hdf5_path}")
                with pd.HDFStore(str(hdf5_path), mode='r') as store:
                    for key in store.keys():
                        clean_key = key.lstrip('/').replace('_', '-')
                        datasets[clean_key] = store[key]
                print(f"âœ… Loaded {len(datasets)} datasets from HDF5")
                return datasets
        except Exception as e:
            print(f"âš ï¸  HDF5 loading failed: {e}")
    
    # Try Parquet
    if preferred_format == 'parquet' or not datasets:
        try:
            parquet_dir = base_path / "parquet_datasets"
            if parquet_dir.exists():
                print(f"ðŸ“‚ Loading from Parquet: {parquet_dir}")
                for parquet_file in parquet_dir.glob("*.parquet"):
                    key = parquet_file.stem.replace('_', '-')
                    datasets[key] = pd.read_parquet(parquet_file)
                print(f"âœ… Loaded {len(datasets)} datasets from Parquet")
                if datasets:
                    return datasets
        except Exception as e:
            print(f"âš ï¸  Parquet loading failed: {e}")
    
    # Try CSV + JSON
    if preferred_format == 'csv' or not datasets:
        try:
            csv_dir = base_path / "csv_datasets"
            metadata_path = csv_dir / "datasets_metadata.json"
            
            if metadata_path.exists():
                print(f"ðŸ“‚ Loading from CSV + JSON: {csv_dir}")
                
                with open(metadata_path, 'r') as f:
                    metadata = json.load(f)
                
                for key, info in metadata['datasets'].items():
                    csv_path = csv_dir / info['filename']
                    if csv_path.exists():
                        datasets[key] = pd.read_csv(csv_path, index_col=0)
                
                print(f"âœ… Loaded {len(datasets)} datasets from CSV + JSON")
                return datasets
        except Exception as e:
            print(f"âš ï¸  CSV + JSON loading failed: {e}")
    
    print("âŒ No datasets could be loaded from alternative formats")
    return {}

# Demonstrate alternative formats if datasets are available
if 'datasets' in globals() and datasets:
    print("ðŸš€ CROSS-PLATFORM SERIALIZATION DEMONSTRATION")
    print("=" * 60)
    
    # Save in multiple formats
    saved_files = save_datasets_alternative_formats(datasets, f'{data_root}/alternative_formats')
    
    print(f"\nðŸ“‹ SAVED FILES SUMMARY:")
    print("=" * 40)
    for format_name, files in saved_files.items():
        print(f"ðŸ“‚ {format_name.upper()}:")
        if isinstance(files, list):
            print(f"   â€¢ {len(files)} files")
        elif isinstance(files, dict):
            print(f"   â€¢ Data files: {len(files.get('data_files', []))}")
            print(f"   â€¢ Metadata: {files.get('metadata_file', 'N/A')}")
        else:
            print(f"   â€¢ {files}")
    
    print(f"\nðŸ’¡ RECOMMENDATIONS FOR YOUR COWORKER:")
    print("=" * 50)
    print("1. ðŸ¥‡ BEST OPTION - HDF5:")
    print("   â€¢ Excellent cross-platform compatibility")
    print("   â€¢ Fast loading/saving")
    print("   â€¢ Preserves data types")
    print("   â€¢ Single file")
    print("")
    print("2. ðŸ¥ˆ SECOND CHOICE - Parquet:")
    print("   â€¢ Great compression")
    print("   â€¢ Fast columnar format")
    print("   â€¢ Multiple files (one per dataset)")
    print("")
    print("3. ðŸ¥‰ FALLBACK - CSV + JSON:")
    print("   â€¢ Universal compatibility")
    print("   â€¢ Human readable")
    print("   â€¢ Larger file sizes")
    print("")
    print("ðŸ“§ Send your coworker the entire 'alternative_formats' folder!")
else:
    print("âš ï¸  No datasets available for alternative format demonstration")
````

````python
# Progress tracking function for imputation across multiple missing mechanisms
def run_imputation_with_progress(datasets_dict, imputation_methods,target_column='target'):
    """
    Run multiple imputation methods with progress tracking across different missing mechanisms.
    
    Args:
        datasets_dict: Dictionary with keys like 'MCAR', 'MAR', 'MNAR' and DataFrame values
        imputation_methods: List of (method_name, method_function) tuples
    """
    results = {}
    total_tasks = len(datasets_dict) * len(imputation_methods)
    
    print(f"ðŸš€ Starting comprehensive imputation pipeline...")
    print(f"ðŸ“Š Missing mechanisms: {len(datasets_dict)} ({', '.join(datasets_dict.keys())})")
    print(f"ðŸ”§ Imputation methods: {len(imputation_methods)} ({', '.join([m[0] for m in imputation_methods])})")
    print(f"ðŸ“ˆ Total tasks: {total_tasks}")
    print("=" * 70)
    
    overall_start = time.time()
    task_times = []
    
    with tqdm(total=total_tasks, desc="Overall Progress") as pbar:
        for mechanism_name, dataset in datasets_dict.items():
            print(f"\nðŸ”¬ Processing {mechanism_name} mechanism...")
            results[mechanism_name] = {}
            
            # Separate features and target for this dataset
            if target_column in dataset.columns:
                X_data = dataset.drop(columns=[target_column])
                y_data = dataset[target_column]
            else:
                X_data = dataset
                y_data = None
            
            original_missing = X_data.isnull().sum().sum()
            print(f"   ðŸ“Š Dataset shape: {X_data.shape}")
            print(f"   ðŸŽ¯ Missing values: {original_missing}")
            
            for method_name, method_func in imputation_methods:
                task_start = time.time()
                
                # Update progress description
                pbar.set_description(f"{mechanism_name} + {method_name}")
                
                try:
                    # Run the imputation method
                    if method_name == 'KNN':
                        # Create mask for KNN - where True means missing
                        mask = X_data.isnull().to_numpy()
                        imputed_data = method_func(X_data, mask)
                    else:
                        imputed_data = method_func(X_data)
                    
                    task_end = time.time()
                    task_time = task_end - task_start
                    task_times.append(task_time)
                    
                    # Validate imputation
                    missing_after = imputed_data.isnull().sum().sum()
                    completion_rate = ((original_missing - missing_after) / original_missing * 100) if original_missing > 0 else 100
                    
                    # Store results
                    results[mechanism_name][method_name] = {
                        'data': imputed_data,
                        'time': task_time,
                        'missing_before': original_missing,
                        'missing_after': missing_after,
                        'completion_rate': completion_rate,
                        'target_data': y_data  # Store target separately
                    }
                    
                    # Calculate and show time estimates
                    avg_time = np.mean(task_times)
                    remaining_tasks = total_tasks - len(task_times)
                    eta_seconds = avg_time * remaining_tasks
                    eta_minutes = eta_seconds / 60
                    
                    pbar.set_postfix({
                        'Mechanism': mechanism_name,
                        'Method': method_name,
                        'Task_Time': f'{task_time:.1f}s',
                        'Completion': f'{completion_rate:.1f}%',
                        'ETA': f'{eta_minutes:.1f}min' if eta_minutes > 1 else f'{eta_seconds:.0f}s'
                    })
                    
                    print(f"     âœ… {method_name}: {completion_rate:.1f}% complete in {task_time:.1f}s")
                    
                except Exception as e:
                    print(f"     âŒ Error in {method_name}: {str(e)}")
                    results[mechanism_name][method_name] = {
                        'error': str(e),
                        'missing_before': original_missing,
                        'target_data': y_data
                    }
                
                pbar.update(1)
    
    total_time = time.time() - overall_start
    print(f"\nâœ… ALL IMPUTATION TASKS COMPLETED!")
    print(f"â±ï¸  Total time: {total_time/60:.1f} minutes")
    print(f"ðŸ“Š Average time per task: {np.mean(task_times):.1f} seconds")
    
    # Summary by mechanism
    print(f"\nðŸ“‹ RESULTS BY MISSING MECHANISM:")
    print("="*50)
    for mechanism, methods in results.items():
        print(f"\nðŸ”¬ {mechanism}:")
        successful_methods = [m for m, r in methods.items() if 'error' not in r]
        failed_methods = [m for m, r in methods.items() if 'error' in r]
        
        print(f"   âœ… Successful: {len(successful_methods)}/{len(methods)} methods")
        if successful_methods:
            avg_completion = np.mean([results[mechanism][m]['completion_rate'] for m in successful_methods])
            print(f"   ðŸ“ˆ Average completion rate: {avg_completion:.1f}%")
        if failed_methods:
            print(f"   âŒ Failed methods: {', '.join(failed_methods)}")
    
    return results

print("âœ… Multi-mechanism progress tracking function ready!")
print("ðŸ”¬ Enhanced for MCAR, MAR, MNAR analysis!")
print("ðŸš€ Ready to run comprehensive imputation comparison!")
````
