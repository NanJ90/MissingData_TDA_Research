```python
data_dir = "/{ROOT_PATH}/climate_missing_data"
```

```python
import pandas as pd
import numpy as np
import os

def load_data(file_path,index_col=None,parse_dates=True):
	"""
	Load data from a CSV file and return a DataFrame.
	"""
	if not os.path.exists(file_path):
		raise FileNotFoundError(f"File {file_path} does not exist.")
	
	df = pd.read_csv(file_path, index_col=index_col, parse_dates=parse_dates)
	return df

dfs = {}

for file in os.listdir(data_dir):
	if file.endswith(".csv"):
		file_path = os.path.join(data_dir, file)
		missing_mech = file.split("_")[-1]
		print(missing_mech)
		try:
			df = load_data(file_path,index_col=0)
			dfs[file] = df
		except Exception as e:
			print(f"Error loading {file}: {e}")
```

```python
dfs.keys()  # Display the keys of the loaded DataFrames
print(dfs['climate_mar.csv'])
#export df in dfs to csv when the value is NaN conver to np.nan
for key, df in dfs.items():
	df.replace({None: np.nan}, inplace=True)
	output_file = os.path.join(data_dir, f"formatted_{key}")
	df.to_csv(output_file)
	print(f"Exported {key} to {output_file}")
```
